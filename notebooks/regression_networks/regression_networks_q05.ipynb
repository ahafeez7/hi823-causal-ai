```python
import pandas as pd
import numpy as np
```

```python
data = {'D': {'M': 7, 'N': 21, 'U': 7, 'V': 0},
        'M': {'D': 52, 'N': 34, 'U': 0, 'V': 7},
        'N': {'D': 24, 'M': 16, 'U': 7, 'V': 30},
        'U': {'D': 38, 'M': 0 , 'N': 5, 'V': 14},
        'V': {'D': 45, 'M': 27, 'N': 3, 'U': 52,}}
```

```python
for key1 in data.keys():
    for key2, value in data[key1].items():
        print(f"{key1}\t{key2}\t{value}")
```

```python
diseases = ['D', 'M', 'N', 'U', 'V']
```

```python
point_diff = pd.DataFrame(index=diseases, columns=diseases)
```

```python
for d1 in diseases:
    for d2 in diseases:
        if d1 == d2:
            point_diff.loc[d1, d2] = 0
        elif d1 in data and d2 in data[d1] and d2 in data and d1 in data[d2]:
            diff = data[d1][d2] - data[d2][d1]
            point_diff.loc[d1, d2] = diff
            point_diff.loc[d2, d1] = -diff
            point_diff.replace(0, np.nan, inplace=True)
```

```python
point_diff_list = point_diff.to_numpy().tolist()
print(point_diff)
```

```python
avg_diff = point_diff.mean(axis=1)
for disease in diseases:
    print(f"AVG OF POINT differentials FOR {disease}: {avg_diff[disease]}")
```

```python
avg_diff = point_diff.mean(axis=1)

ranked_diff = avg_diff.sort_values().reset_index()
ranked_diff.columns = ['Disease', 'Avg. Point Differential']
ranked_diff.index = np.arange(1, len(ranked_diff) + 1)

print(ranked_diff)
```
