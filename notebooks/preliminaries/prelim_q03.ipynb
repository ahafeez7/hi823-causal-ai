# DAG Analysis

This notebook analyzes the given graph and answers:

- Parents, ancestors, descendants
- Simple paths and directed paths
- DAG verification

```python
# Import library
import networkx as nx
```

```python
# Create directed graph
G = nx.DiGraph()

# Add edges based on the diagram
G.add_edges_from([
    ("X", "Y"),
    ("X", "W"),
    ("W", "Y"),
    ("W", "Z"),
    ("Y", "Z"),
    ("Y", "T"),
    ("Z", "T")
])

# Display edges
list(G.edges())
```

```python
# Variables that precede Z (ancestors of Z)
precede_Z = nx.ancestors(G, "Z")
print("Variables that precede Z:", precede_Z)
```

```python
# Parents of Z
parents_Z = list(G.predecessors("Z"))
print("Parents of Z:", parents_Z)
```

```python
# Ancestors of Z
ancestors_Z = nx.ancestors(G, "Z")
print("Ancestors of Z:", ancestors_Z)
```

```python
# Children of W
children_W = list(G.successors("W"))
print("Children of W:", children_W)
```

```python
# Descendants of W
descendants_W = nx.descendants(G, "W")
print("Descendants of W:", descendants_W)
```

```python
# All simple paths between X and T
simple_paths = list(nx.all_simple_paths(G, source="X", target="T"))

print("Simple paths from X to T:")
for path in simple_paths:
    print(" -> ".join(path))
```
