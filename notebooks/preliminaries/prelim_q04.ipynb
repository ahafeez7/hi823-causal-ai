# Networks from Independence Assumptions

This notebook constructs causal networks based on independence assumptions
and prints the corresponding prediction formulas.`

```python
# Install dependency (run once)
!pip install networkx
```

```python
# Import library
import networkx as nx
```

```python
# Case 1: I(X, Y)

G1 = nx.DiGraph()
G1.add_edges_from([("X", "Z"), ("Y", "Z")])

print("Case 1 Network Edges:", list(G1.edges()))
print("Prediction formula: P(Z | X, Y)")
```

```python
# Case 2: I(X, Y), Not I(X, Y | Z)

G2 = nx.DiGraph()
G2.add_edges_from([("X", "Z"), ("Y", "Z")])

print("Case 2 Network Edges:", list(G2.edges()))
print("Prediction formula: P(Z | X, Y)")
```

```python
# Case 3: I(X, Y), I(X, Y | Z), Y measured last

G3 = nx.DiGraph()
G3.add_edges_from([("Z", "X"), ("Z", "Y")])

print("Case 3 Network Edges:", list(G3.edges()))
print("Prediction formula: P(Y | Z)")
```

```python
# Case 4: I(X, Y), I(X, Y | Z), I({X,Y}, W | Z), W measured last

G4 = nx.DiGraph()
G4.add_edges_from([("Z", "X"), ("Z", "Y"), ("Z", "W")])

print("Case 4 Network Edges:", list(G4.edges()))
print("Prediction formula: P(W | Z)")
```

```python
# Case 5: I(X, Y), I(Z, W), ordering constraints

G5 = nx.DiGraph()
G5.add_edges_from([("X", "Z"), ("Y", "W")])

print("Case 5 Network Edges:", list(G5.edges()))
print("Prediction formula: P(W | Y)")
```
