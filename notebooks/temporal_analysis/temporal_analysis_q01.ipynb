```python
#Define Nodes - same order as your variable first
import networkx as nx
import matplotlib.pyplot as plt

# create directed graph
G = nx.DiGraph()

# nodes in causal order
nodes = [

# demographics
"Age",
"Female",

# symptoms
"Shivering",
"Fatigue",
"Loss_of_Taste",
"Fever",
"Headaches",
"Loss_of_Smell",
"Chills",
"Muscle_Aches",
"Diarrhea",
"Cough",
"Shortness_of_Breath",
"Runny_Nose",
"Sore_Throat",
"Loss_Balance",
"Vomiting",
"Joint_Pain",
"Loss_of_Appetite",
"Difficulty_Breathing",
"Excess_Sweating",

# outcome
"COVID_Test_Result"
]

G.add_nodes_from(nodes)
```

```python
# add directed/causal edges
edges = [

("Age","Shivering"),
("Age","Fatigue"),
("Female","Loss_of_Taste"),

("Shivering","Fever"),
("Fatigue","Fever"),
("Loss_of_Taste","Headaches"),

("Fever","Chills"),
("Fever","Loss_of_Smell"),
("Headaches","Muscle_Aches"),

("Runny_Nose","Cough"),
("Sore_Throat","Cough"),

("Cough","Shortness_of_Breath"),
("Shortness_of_Breath","Difficulty_Breathing"),

("Vomiting","Loss_of_Appetite"),
("Muscle_Aches","Joint_Pain"),

("Loss_of_Smell","COVID_Test_Result"),
("Chills","COVID_Test_Result"),
("Muscle_Aches","COVID_Test_Result"),
("Cough","COVID_Test_Result"),
("Shortness_of_Breath","COVID_Test_Result"),
("Loss_of_Appetite","COVID_Test_Result"),
("Excess_Sweating","COVID_Test_Result")
]

G.add_edges_from(edges)
```

```python
#Node colros - by causal type
node_colors = {

# demographics
"Age":"lightblue",
"Female":"lightblue",

# layer 2
"Shivering":"gold",
"Fatigue":"gold",
"Loss_of_Taste":"gold",

# layer 3
"Fever":"khaki",
"Headaches":"khaki",
"Diarrhea":"khaki",
"Runny_Nose":"khaki",
"Sore_Throat":"khaki",
"Loss_Balance":"khaki",
"Vomiting":"khaki",
"Joint_Pain":"khaki",
"Difficulty_Breathing":"khaki",

# layer 4
"Loss_of_Smell":"lightgreen",
"Chills":"lightgreen",
"Muscle_Aches":"lightgreen",
"Cough":"lightgreen",
"Shortness_of_Breath":"lightgreen",
"Loss_of_Appetite":"lightgreen",
"Excess_Sweating":"lightgreen",

# outcome
"COVID_Test_Result":"lightgray"
}
```

```python
#Node positions - layered layout
pos = {

# Layer 1
"Age":(-4,4),
"Female":(-4,2),

# Layer 2
"Shivering":(-2,3),
"Fatigue":(-1,3),
"Loss_of_Taste":(0,3),

# Layer 3
"Fever":(-3,1.5),
"Headaches":(-2,1.5),
"Diarrhea":(-1,1.5),
"Runny_Nose":(0,1.5),
"Sore_Throat":(1,1.5),
"Loss_Balance":(2,1.5),
"Vomiting":(3,1.5),
"Joint_Pain":(4,1.5),
"Difficulty_Breathing":(5,1.5),

# Layer 4
"Loss_of_Smell":(-2,0),
"Chills":(-1,0),
"Muscle_Aches":(0,0),
"Cough":(1,0),
"Shortness_of_Breath":(2,0),
"Loss_of_Appetite":(3,0),
"Excess_Sweating":(4,0),

# Layer 5
"COVID_Test_Result":(6,0.5)
}
```

```python
#Draw the graph
plt.figure(figsize=(14,8))

# draw nodes
nx.draw_networkx_nodes(
    G,
    pos,
    node_size=2800,
    node_color=[node_colors.get(n, "lightgray") for n in G.nodes()],
    edgecolors="black",
    linewidths=1.5
)

# draw edges
nx.draw_networkx_edges(
    G,
    pos,
    arrows=True,
    arrowstyle="->",
    arrowsize=20,
    width=1.5
)

# draw labels
nx.draw_networkx_labels(
    G,
    pos,
    font_size=10,
    font_weight="bold"
)

plt.title("COVID Symptom Causal Network", fontsize=14)

plt.axis("off")
plt.tight_layout()

plt.show()
```
