```python
from graphviz import Digraph

#What is the probability of taking medications given that the person is female

# Create graph
g = Digraph()

# Node style simlar to Netica graph/node
g.attr('node', shape='box', style='filled', fillcolor='lightyellow')

# Nodes with labels
g.node('Gender', 'Gender\n\nMale 0\nFemale 100')
g.node('TookMeds', 'TookMeds\n\nYes 73.0\nNo 27.0')
g.node('RemOfSym', 'RemOfSym\n\nYes 71.9\nNo 28.1')

# Edges
g.edge('Gender', 'TookMeds')
g.edge('Gender', 'RemOfSym')
g.edge('TookMeds', 'RemOfSym')

# Display
g
```
